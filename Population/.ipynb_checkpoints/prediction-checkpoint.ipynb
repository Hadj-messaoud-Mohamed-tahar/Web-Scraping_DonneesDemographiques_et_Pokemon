```python
# Importation des bibliothèques nécessaires
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

# URL du site Web
url = "https://www.worldometers.info/world-population/population-by-country/"

# Démarrer le chronomètre
start_time = time.time()

# Envoi de la requête HTTP et analyse avec BeautifulSoup
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Trouver la première table sur la page
table = soup.find('table')

# Vérification et extraction des données
rows = table.find_all('tr')

# Extraire les données ligne par ligne
data = []
for row in rows[1:]:  # Ignorer la première ligne (en-têtes)
    cols = row.find_all('td')
    row_data = [col.text.strip() for col in cols]
    
    # Ajouter le lien associé au pays
    link = cols[1].find("a")  # La deuxième colonne contient le pays
    href = link.get("href") if link else None
    row_data.append(href)  # Ajouter le lien à la fin de la ligne
    data.append(row_data)

# Définir les noms de colonnes 
columns = [
    'Rank', 'Country', 'Population', 'Yearly Change', 'Net Change',
    'Density (P/Km²)', 'Land Area (Km²)', 'Migrants (net)', 'Fertility Rate',
    'Median Age', 'Urban Pop %', 'World Share', 'Link'
]

# Créer un DataFrame
df = pd.DataFrame(data, columns=columns)

# Fonction pour convertir en numérique de manière sûre
def safe_convert_to_numeric(value):
    try:
        return pd.to_numeric(value.replace(',', '').replace('%', ''), errors='coerce')
    except AttributeError:
        return np.nan

# Colonnes numériques à nettoyer
numeric_columns = [
    'Population', 'Yearly Change', 'Net Change', 'Density (P/Km²)', 
    'Land Area (Km²)', 'Migrants (net)', 'Fertility Rate', 'Median Age', 
    'Urban Pop %', 'World Share'
]

# Appliquer la conversion à toutes les colonnes concernées
for col in numeric_columns:
    df[col] = df[col].apply(safe_convert_to_numeric)

# Sauvegarder les données nettoyées dans un fichier CSV
csv_path = 'population_by_country.csv'
df.to_csv(csv_path, index=False)

# Afficher un aperçu des premières lignes du DataFrame
print(f"Scraping complet! Données sauvegardées dans {csv_path}. Temps pris: {time.time() - start_time:.2f} secondes.")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisation des 10 pays les plus peuplés
top_10 = df[['Country', 'Population']].sort_values(by='Population', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Population', y='Country', data=top_10, palette="viridis")
plt.title("Top 10 des pays les plus peuplés")
plt.xlabel("Population (en milliards)")
plt.ylabel("Pays")
plt.show()
```

```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Initialiser le WebDriver
driver = webdriver.Chrome()

# URL de base
v = "https://www.worldometers.info"

# Assurez-vous que df contient bien les liens dans la colonne 'Link'
urls_list = df["Link"].tolist()

# Liste pour stocker les données extraites
data1 = []
g = 0  

# Liste des colonnes 
columns1 = ['Country', 'Year', 'Population', 'Yearly %  Change', 'Yearly Change', 'Migrants (net)', 'Median Age', 
            'Fertility Rate', 'Density (P/Km²)', 'Urban Pop %', 'Urban Population', "Country's Share of World Pop", 
            'World Population', 'Global Rank']

# Parcourir chaque URL
for url in urls_list:
    base_url = v + url
    print(f"Traitement de l'URL : {base_url}")
    start_time = time.time()

    # Utiliser Selenium pour ouvrir l'URL
    driver.get(base_url)
    
    # Attendre un peu pour s'assurer que la page est entièrement chargée
    time.sleep(3)

    # Récupérer le contenu de la page avec BeautifulSoup
    soup1 = BeautifulSoup(driver.page_source, 'html.parser')

    # Trouver les tables
    tables1 = soup1.find_all('table')
    if not tables1:
        print(f"Aucune table trouvée sur {base_url}")
        continue

    print(f"{len(tables1)} table(s) trouvée(s)")

    # On suppose que la table cible est celle qui nous intéresse
    table1 = tables1[3]
    
    # Extraire les en-têtes (<th>)
    headers = [th.text.strip().replace('<br>', ' ') for th in table1.find_all('th')]
    headers = headers[:-1] + ['Global Rank']  # Remplacer le dernier en-tête avec "Global Rank"

    # Vérifier l'ordre et la correspondance des colonnes
    header_indices = {col: (headers.index(col) if col in headers else -1) for col in columns1[1:]}  # Exclure "Country"

    rows1 = table1.find_all('tr')
    print(f"Nombre de lignes dans la table : {len(rows1)}")

    # Parcourir les lignes de la table
    for row in rows1[1:]:  # Ignorer la première ligne (en-têtes)
        row_data1 = []  # Réinitialiser les données pour chaque ligne

        # Ajouter le nom du pays depuis df
        try:
            country_name = df["Country"].iloc[g]  # Assurez-vous que g est valide
        except IndexError:
            print(f"Index {g} hors des limites pour 'Country'")
            break
        row_data1.append(country_name)

        # Ajouter les données des colonnes dans le bon ordre
        cols = row.find_all('td')
        temp_data = [col.text.strip() for col in cols]
        ordered_row = []

        for col_name in columns1[1:]:
            if header_indices[col_name] != -1 and header_indices[col_name] < len(temp_data):
                ordered_row.append(temp_data[header_indices[col_name]])
            else:
                ordered_row.append("")  # Ajouter une valeur vide si la colonne est absente

        row_data1.extend(ordered_row)
        data1.append(row_data1)

    # Passer au prochain pays
    g += 1

# Fermer le navigateur
driver.quit()

# Créer un DataFrame à partir des données extraites
final_df = pd.DataFrame(data1, columns=columns1)
print(final_df.head())
```
